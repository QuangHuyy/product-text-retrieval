```python
import pandas as pd
df =  pd.read_csv("../tdgd_laptop/laptops_all.csv")

df.head()
```

```python
# related_cols = {
#     'laptop': ['ram', 'cpu'],
#     'printer': ['speed', 'production']
# }


related_cols_w_weight = ['ram', 'cpu1'] #Cột mà mình sẽ dùng để tìm ra những sản phẩm tương đương
found_keys = {} #Những giá trị của cột
for c in related_cols_w_weight:
    if c not in found_keys:
        found_keys[c] = {}
```

```python
for index, row in df.iterrows():
    for c in related_cols_w_weight:
        #row[c]: giá trị của cột đó
        if row[c] is not None:
            if row[c] in found_keys[c].keys():
                found_keys[c][row[c]].append(row['code']) #Thêm mã sản phẩm này
            #Nếu chưa thấy sản phẩm với key này thì thêm list mới
            else:
                found_keys[c][row[c]] = [row['code']]
    
print(found_keys)
```

```python
class weighted_attr:
    def __init__(self, attr, weight):
        self.attr = attr
        self.weight = weight
    def __repr__(self):
        return str(self.attr) + ", weight: " + str(self.weight)

def get_index_by_value(list, value):
    indices = [i for i, v in enumerate(list) if v.attr == value]
    if len(indices) > 1:
        raise RuntimeError('found more indexes 1')
    if len(indices) == 0:
        return None
        # raise RuntimeError('found no indexes')
    return indices[0]

laptop_weighted_attr = []
laptop_weighted_attr.append(weighted_attr('ram', 3))
laptop_weighted_attr.append(weighted_attr('cpu1', 3))

relevant_items = {}
for index, row in df.iterrows():
    relevant_items[row['code']] = []
    for wa in laptop_weighted_attr:
        #Danh sách laptops liên quan
        laptops_by_attribute = found_keys[wa.attr][row[wa.attr]]
        

        #Thêm vào mảng
        #Nếu laptop với code này chưa có trong mảng thì add vô, else + thêm weight
        for l in laptops_by_attribute:
            # if l == row['code']:
            #     continue
            index = get_index_by_value(relevant_items[row['code']], l)
            if index is None:
                relevant_items[row['code']].append(weighted_attr(l, wa.weight))
            else:
                relevant_items[row['code']][index].weight += wa.weight
```
