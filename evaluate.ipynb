```python
eval_dict = {
    'f_score': [],
    'precision': [],
    'recall': [],
    'query': [],
    'result': [],
    'ground_truth': [],
}

for index, row in e_df.iterrows():
    result = s.evaluate_query(row['query'])
    g_truth = row['result'].split()
    
    join_result_truth = [x for x in g_truth if x in result]
    precision = len(join_result_truth) / len(result) if len(result) != 0 else 0
    recall = len(join_result_truth) / len(g_truth)
    f_score = (2*precision*recall)/(precision+recall) if precision+recall != 0 else 0

    print(f_score)
    
    eval_dict['f_score'].append(f_score)
    eval_dict['precision'].append(precision)
    eval_dict['recall'].append(recall)
    eval_dict['query'].append(row['query'])
    eval_dict['result'].append(result)
    eval_dict['ground_truth'].append(g_truth)    
    
# print(eval_dict)
```

```python
statistics = ['f_score', 'precision', 'recall']

for s in statistics:
    print(f"{s}: ", sum(eval_dict[s])/len(eval_dict[s]))
```

```python
eval_rs_df = pd.DataFrame(eval_dict)
eval_rs_df.to_csv("evaluate_result.csv", index=False)
```
